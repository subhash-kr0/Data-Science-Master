```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import pandas as pd

# Set up the WebDriver (replace 'path/to/chromedriver' with the actual path)
driver = webdriver.Chrome(executable_path='path/to/chromedriver')
driver.get("https://www.youtube.com/@PW-Foundation/videos")

# Give the page some time to load
time.sleep(5)
```

```python
# Extract data
videos = driver.find_elements(By.XPATH, '//*[@id="video-title"]')[:5]
thumbnails = driver.find_elements(By.XPATH, '//*[@id="img"][@width="360"]')[:5]
views = driver.find_elements(By.XPATH, '//*[@id="metadata-line"]/span[1]')[:5]
post_times = driver.find_elements(By.XPATH, '//*[@id="metadata-line"]/span[2]')[:5]

# Initialize lists to hold data
video_urls = []
thumbnail_urls = []
titles = []
view_counts = []
post_times_list = []

# Loop through the first 5 videos to extract information
for i in range(5):
    video_urls.append(videos[i].get_attribute('href'))
    thumbnail_urls.append(thumbnails[i].get_attribute('src'))
    titles.append(videos[i].get_attribute('title'))
    view_counts.append(views[i].text)
    post_times_list.append(post_times[i].text)

# Close the driver
driver.quit()

# Create a DataFrame
data = {
    'Title': titles,
    'Video URL': video_urls,
    'Thumbnail URL': thumbnail_urls,
    'Views': view_counts,
    'Posted': post_times_list
}

df = pd.DataFrame(data)

# Save to CSV
df.to_csv('youtube_videos.csv', index=False)
print("Data saved to youtube_videos.csv")
```
